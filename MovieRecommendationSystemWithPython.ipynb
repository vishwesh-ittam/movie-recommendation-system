Importing the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-processing

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

Selecting relevant features for recommendation

```python
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#Replacing null values with null strings
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#Combining all the five relevant selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#Converting the text data with feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
#Getting a similarity score using cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#Getting user input
movie_name = input('Enter your favourite movie name: ')
```

```python
#Creating a list of all movie titles
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#Finding the closest match of the user's movie title in the dataset - He might mispell the title
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
closest_match=find_close_match[0]
print(closest_match)
```

```python
#Finding the index of the movie title
closest_match = find_close_match[0]
index_of_the_movie = movies_data[movies_data.title == closest_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#Getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
len(similarity_score)
```

```python
#Sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)#Basically in the list of similarity_score we are sorting the x[1] elements in descending order
print(sorted_similar_movies)
```

```python
#print the name of similar movies on the index
print('Movies suggested for you : \n')

i=1

for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.' , title_from_index)
    i+=1
```

```python
movie_name = input('Enter your favourite movie name: ')

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

if len(find_close_match) == 0:
    print(f'No close match found for "{movie_name}". Please try another movie.')
else:
    closest_match = find_close_match[0]
    index_of_the_movie = movies_data[movies_data.title == closest_match]['index'].values[0]

    similarity_score = list(enumerate(similarity[index_of_the_movie]))

    sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

    print('Movies suggested for you : \n')

    i=1

    for movie in sorted_similar_movies:
        index=movie[0]
        title_from_index = movies_data[movies_data.index==index]['title'].values[0]
        if (i<30):
            print(i, '.' , title_from_index)
            i+=1
```
